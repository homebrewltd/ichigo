```python
OLD_AUDIO_PATH = "/home/root/Workspace/synthetic_data_generation/sound_instruct_llama3/data/old_audio/audio_dataset"
```

```python
# load each arrow file and concatenate them
import os
from datasets import Dataset, concatenate_datasets

ds = []
for file in os.listdir(OLD_AUDIO_PATH):
    if file.split(".")[-1] == "arrow":
        dataset = Dataset.from_file(os.path.join(OLD_AUDIO_PATH, file))
        ds.append(dataset)

ds
```

```python
ds = concatenate_datasets(ds)
ds
```

```python
ds[0]['audio']
```

```python
# save the concatenated dataset into wav files
import soundfile as sf
from tqdm.notebook import tqdm
error_idx = []
def save_wav(ds,
             output_dir="/home/root/Workspace/synthetic_data_generation/sound_instruct_llama3/data/audio"):
    for i, row in tqdm(enumerate(ds)):
        try:
            item = row['audio']
            path, array, sample_rate = item["path"], item["array"], item["sampling_rate"]
            filename = os.path.join(output_dir, path)
            sf.write(filename, array, sample_rate)
        except Exception as e:
            print(f"Error at index {i}: {e}")
            error_idx.append(i)
            continue

save_wav(ds)
```
